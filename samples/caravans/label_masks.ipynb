This notebook takes the masks which were created in the making_training_data.ipynb and makes labels out of them using the label function from skimage.measure. These are the labels that you will want to feed to the model.

```python
from skimage.measure import label
from skimage.io import imread, imsave
import os 
from glob import glob
from numpy import unique
import warnings
```

```python
warnings.simplefilter("ignore")
```

```python
def labeller(image_path):
    image = imread(image_path)
    last_part = os.path.basename(os.path.normpath(image_path))
    print(last_part)
    num = last_part.split("_")[1]
    print(num)
    labelled = label(image)
    outfolder = ".../caravanProject/imagery/images_maia_topo_polygons/training_patches/labels/{}.png".format(num)
    imsave(outfolder, labelled)
    
```

```python
#goes through your folder of masks and labels & writes them to a new folder 
path = ".../caravanProject/imagery/images_maia_topo_polygons/training_patches/Masks"
for i, image in enumerate(glob(path+"\*.tif")):
    labeller(image)
```
